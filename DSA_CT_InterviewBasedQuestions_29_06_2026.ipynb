### [Stack] Problem 402: Remove K Digits

**Algorithm & Technique:** Monotonic Stack

**Logic:**
Step 1: Create a stack to store digits in increasing order.

Step 2: Iterate through each digit of the number.

Step 3: While the current digit is smaller than the stack's top and $k > 0$, pop from the stack and decrement $k$.

Step 4: Push the current digit onto the stack.

Step 5: After the loop, if $k > 0$, remove remaining digits from the end of the stack.

Step 6: Join the stack into a string, remove leading zeros, and return the result (or "0" if empty).

```python
class Solution:
    def removeKdigits(self, num: str, k: int) -> str:
        stack = []
        for digit in num:
            while stack and k > 0 and stack[-1] > digit:
                stack.pop()
                k -= 1
            stack.append(digit)

        while k > 0:
            stack.pop()
            k -= 1

        result = "".join(stack).lstrip('0')

        return result if result else "0"

sol = Solution()
print(f"'1432219', k=3 -> {sol.removeKdigits('1432219', 3)}")
print(f"'10200', k=1 -> {sol.removeKdigits('10200', 1)}")
print(f"'10', k=2 -> {sol.removeKdigits('10', 2)}")
print(f"'12345', k=2 -> {sol.removeKdigits('12345', 2)}")
print(f"'4321', k=1 -> {sol.removeKdigits('4321', 1)}")
```

### [Stack] Problem 84: Largest Rectangle in Histogram

**Algorithm & Technique:** Monotonic Stack

**Logic:**
Step 1: Initialize an empty stack to store indices of the bars.

Step 2: Iterate through the histogram bars.

Step 3: If the current bar is shorter than the bar at the stack's top index, pop from the stack and calculate the area using the popped height.

Step 4: Determine the width as the distance between the current index and the new stack top (or start of the histogram).

Step 5: Update the maximum area found so far.

Step 6: Repeat the calculation for any remaining bars in the stack after the iteration.

```python
class Solution:
    def largestRectangleArea(self, heights: list[int]) -> int:
        max_area = 0
        stack = []

        for i, h in enumerate(heights):
            start_index = i
            while stack and stack[-1][1] > h:
                idx, height = stack.pop()
                max_area = max(max_area, height * (i - idx))
                start_index = idx
            stack.append((start_index, h))

        for i, h in stack:
            max_area = max(max_area, h * (len(heights) - i))

        return max_area

sol = Solution()
print(f"[2,1,5,6,2,3] -> {sol.largestRectangleArea([2,1,5,6,2,3])}")
print(f"[2,4] -> {sol.largestRectangleArea([2,4])}")
print(f"[0,9] -> {sol.largestRectangleArea([0,9])}")
print(f"[6,7,5,2,4,5,9,3] -> {sol.largestRectangleArea([6,7,5,2,4,5,9,3])}")
```

### [Queue / Monotonic Queue] Problem 862: Shortest Subarray with Sum at Least K

**Algorithm & Technique:** Prefix Sum + Monotonic Deque

**Logic:**
Step 1: Precompute prefix sums for the input array.

Step 2: Initialize a deque to store indices of prefix sums in increasing order.

Step 3: Iterate through the prefix sums.

Step 4: While the difference between the current prefix sum and the sum at the deque's front is at least $k$, update the minimum length and pop from the front.

Step 5: While the current prefix sum is less than or equal to the sum at the deque's back, pop from the back to maintain monotonicity.

Step 6: Push the current index onto the deque.

```python
import collections

class Solution:
    def shortestSubarray(self, nums: list[int], k: int) -> int:
        n = len(nums)
        prefix_sum = [0] * (n + 1)
        for i in range(n):
            prefix_sum[i + 1] = prefix_sum[i] + nums[i]

        min_len = float('inf')
        deque = collections.deque()

        for i in range(n + 1):
            while deque and prefix_sum[i] - prefix_sum[deque[0]] >= k:
                min_len = min(min_len, i - deque.popleft())

            while deque and prefix_sum[i] <= prefix_sum[deque[-1]]:
                deque.pop()

            deque.append(i)

        return min_len if min_len != float('inf') else -1

sol = Solution()
print(f"nums = [1], k = 1 -> {sol.shortestSubarray([1], 1)}")
print(f"nums = [1,2], k = 4 -> {sol.shortestSubarray([1,2], 4)}")
print(f"nums = [2,-1,2], k = 3 -> {sol.shortestSubarray([2,-1,2], 3)}")
print(f"nums = [-2,1,1,2], k = 3 -> {sol.shortestSubarray([-2,1,1,2], 3)}")
print(f"nums = [84,-37,32,40,95], k = 167 -> {sol.shortestSubarray([84,-37,32,40,95], 167)}")
```

### [Queue / Monotonic Queue] Problem 1425: Constrained Subsequence Sum

**Algorithm & Technique:** DP + Monotonic Deque

**Logic:**
Step 1: Initialize a DP array where `dp[i]` represents the max sum ending at index $i$.

Step 2: Initialize a deque to maintain indices of the largest DP values in the current window of size $k$.

Step 3: For each element, remove indices from the front of the deque that are outside the window $[i-k, i-1]$.

Step 4: Calculate `dp[i]` by adding the current number to the max of 0 and the DP value at the front index of the deque.

Step 5: Maintain the deque in decreasing order by removing indices with smaller DP values before adding the current index.

```python
import collections

class Solution:
    def constrainedSubsetSum(self, nums: list[int], k: int) -> int:
        n = len(nums)
        dp = [0] * n
        max_overall_sum = -float('inf')

        deque = collections.deque()

        for i in range(n):
            while deque and deque[0] < i - k:
                deque.popleft()

            current_val_from_prev = 0
            if deque:
                current_val_from_prev = max(0, dp[deque[0]])

            dp[i] = nums[i] + current_val_from_prev

            max_overall_sum = max(max_overall_sum, dp[i])

            while deque and dp[deque[-1]] <= dp[i]:
                deque.pop()
            deque.append(i)

        return max_overall_sum

# Example Usage with corrected logic:
sol_corrected = Solution()
print(f"nums = [10,2,-10,5,20], k = 2 -> {sol_corrected.constrainedSubsetSum([10,2,-10,5,20], 2)}")
print(f"nums = [-1,-2,-3], k = 1 -> {sol_corrected.constrainedSubsetSum([-1,-2,-3], 1)}")
print(f"nums = [10,-2,-10,-5,20], k = 2 -> {sol_corrected.constrainedSubsetSum([10,-2,-10,-5,20], 2)}")
```

### [Queue / Monotonic Queue] Problem 1696: Jump Game VI

**Algorithm & Technique:** DP + Monotonic Deque

**Logic:**
Step 1: Initialize a DP array where `dp[i]` stores the max score to reach index $i$.

Step 2: Use a deque to store indices of the DP array for a sliding window of size $k$.

Step 3: For each step, pop indices from the front that are more than $k$ steps behind.

Step 4: Set `dp[i]` as the sum of the current value and the DP value at the front of the deque.

Step 5: Pop from the back of the deque while the current `dp[i]` is greater than the DP values of stored indices.

Step 6: Append the current index to the deque and return the final DP value.

```python
import collections

class Solution:
    def maxResult(self, nums: list[int], k: int) -> int:
        n = len(nums)
        dp = [0] * n
        dp[0] = nums[0]

        deque = collections.deque([0])

        for i in range(1, n):
            while deque and deque[0] < i - k:
                deque.popleft()

            dp[i] = nums[i] + dp[deque[0]]

            while deque and dp[deque[-1]] <= dp[i]:
                deque.pop()
            deque.append(i)

        return dp[n - 1]

sol = Solution()
print(f"nums = [1,-1,-2,4,-7,3], k = 2 -> {sol.maxResult([1,-1,-2,4,-7,3], 2)}")
print(f"nums = [10,-5,-2,4,0,3], k = 3 -> {sol.maxResult([10,-5,-2,4,0,3], 3)}")
print(f"nums = [1,-5,-20,4,-1,3,-6,-3], k = 2 -> {sol.maxResult([1,-5,-20,4,-1,3,-6,-3], 2)}")
```

### [Queue / Monotonic Queue] Problem 1429: First Unique Number (Subscribe)

**Algorithm & Technique:** OrderedDict + Hash Set

**Logic:**
Step 1: Create an OrderedDict to track unique elements in the order they were added.

Step 2: Create a set to store every element encountered at least once.

Step 3: When adding a number, check if it has been seen before.

Step 4: If it's a new number, add it to both the dictionary and the set.

Step 5: If it was already in the dictionary, it is no longer unique; remove it.

Step 6: To show the first unique, simply return the first key of the OrderedDict.

```python
import collections

class FirstUnique:

    def __init__(self, nums: list[int]):
        self.unique_elements = collections.OrderedDict()
        self.all_elements_seen = set()

        for num in nums:
            self.add(num)

    def showFirstUnique(self) -> int:
        if self.unique_elements:
            return next(iter(self.unique_elements))
        else:
            return -1

    def add(self, value: int) -> None:
        if value not in self.all_elements_seen:
            self.unique_elements[value] = True
            self.all_elements_seen.add(value)
        elif value in self.unique_elements:
            del self.unique_elements[value]

print("Test Case 1:")
firstUnique = FirstUnique([2,3,5])
print(f"showFirstUnique: {firstUnique.showFirstUnique()}")
firstUnique.add(5)
print(f"showFirstUnique: {firstUnique.showFirstUnique()}")
firstUnique.add(2)
print(f"showFirstUnique: {firstUnique.showFirstUnique()}")
firstUnique.add(3)
print(f"showFirstUnique: {firstUnique.showFirstUnique()}")
```

### [Recursion (Backtracking)] Problem 51: N-Queens

**Algorithm & Technique:** Backtracking

**Logic:**
Step 1: Create sets to track occupied columns and diagonals (positive and negative).

Step 2: Define a recursive function to place queens row by row.

Step 3: In each row, iterate through every column.

Step 4: Check if placing a queen in the current column and diagonals is valid using the sets.

Step 5: If valid, mark the columns/diagonals as occupied and move to the next row.

Step 6: Backtrack by removing the queen and unmarking the constraints to explore other possibilities.

```python
class Solution:
    def solveNQueens(self, n: int) -> list[list[str]]:
        col = set()
        pos_diag = set()
        neg_diag = set()

        board = [['.' for _ in range(n)] for _ in range(n)]
        res = []

        def backtrack(r):
            if r == n:
                copy = ["".join(row) for row in board]
                res.append(copy)
                return

            for c in range(n):
                if c in col or (r + c) in pos_diag or (r - c) in neg_diag:
                    continue

                col.add(c)
                pos_diag.add(r + c)
                neg_diag.add(r - c)
                board[r][c] = 'Q'

                backtrack(r + 1)

                col.remove(c)
                pos_diag.remove(r + c)
                neg_diag.remove(r - c)
                board[r][c] = '.'

        backtrack(0)
        return res

sol = Solution()
print(f"N = 4 -> {sol.solveNQueens(4)}")
print(f"N = 1 -> {sol.solveNQueens(1)}")
```

Step 1: Create a min-heap to store the head nodes of all $k$ linked lists.

Step 2: Initialize a dummy node to act as the head of the merged list.

Step 3: Pop the smallest node from the heap and attach it to the merged list.

Step 4: If the popped node has a next node, push that next node into the heap.

Step 5: Repeat until the heap is empty.

Step 6: Return the merged list starting from `dummy.next`.

```python
import heapq

# Definition for singly-linked list.
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next

class Solution:
    def mergeKLists(self, lists: list[ListNode]) -> ListNode:
        heap = []
        # Add the head of each list to the heap
        for i, l in enumerate(lists):
            if l:
                # Store (value, list_index) to handle duplicate values and fetch next node
                heapq.heappush(heap, (l.val, i))

        dummy = ListNode(0)
        curr = dummy

        while heap:
            val, i = heapq.heappop(heap)
            curr.next = ListNode(val)
            curr = curr.next

            # Move the pointer of the extracted list forward
            lists[i] = lists[i].next
            if lists[i]:
                heapq.heappush(heap, (lists[i].val, i))

        return dummy.next
```

### [Linked List] Problem 25: Reverse Nodes in k-Group

**Algorithm & Technique:** Recursion

**Logic:**
Step 1: Check if there are at least $k$ nodes left in the list; if not, return the head as is.

Step 2: Reverse the first $k$ nodes using the standard iterative reversal approach.

Step 3: After reversing, the original head becomes the tail of the reversed group.

Step 4: Recursively call the function for the remaining part of the list.

Step 5: Connect the tail of the current reversed group to the result of the recursive call.

Step 6: Return the new head of the reversed group.

```python
class Solution:
    def reverseKGroup(self, head: ListNode, k: int) -> ListNode:
        # Check if there are k nodes to reverse
        curr = head
        for _ in range(k):
            if not curr: return head
            curr = curr.next

        # Reverse k nodes
        prev = None
        curr = head
        for _ in range(k):
            nxt = curr.next
            curr.next = prev
            prev = curr
            curr = nxt

        # Recursively call for the rest of the list
        head.next = self.reverseKGroup(curr, k)
        return prev
```

### [Linked List] Problem 138: Copy List with Random Pointer

**Algorithm & Technique:** Hash Map (Dictionary)

**Logic:**
Step 1: Create a hash map to map original nodes to their corresponding newly created copy nodes.

Step 2: Traverse the original list once to create copies of all nodes and store them in the map.

Step 3: Traverse the original list a second time.

Step 4: Use the map to set the `next` and `random` pointers for each copy node based on the original node's pointers.

Step 5: Return the copy of the head node from the map.

```python
class Node:
    def __init__(self, x: int, next: 'Node' = None, random: 'Node' = None):
        self.val = int(x)
        self.next = next
        self.random = random

class Solution:
    def copyRandomList(self, head: 'Node') -> 'Node':
        if not head: return None
        mapping = {}
        curr = head
        while curr:
            mapping[curr] = Node(curr.val)
            curr = curr.next
        curr = head
        while curr:
            mapping[curr].next = mapping.get(curr.next)
            mapping[curr].random = mapping.get(curr.random)
            curr = curr.next
        return mapping[head]
```

### [Linked List] Problem 460: LFU Cache

**Algorithm & Technique:** Doubly Linked List + Hash Map

**Logic:**
Step 1: Use a hash map to store keys and their corresponding node objects.

Step 2: Use another hash map to store frequency counts mapped to Doubly Linked Lists of nodes with that frequency.

Step 3: Track the `min_freq` to quickly find the least frequently used items for eviction.

Step 4: When a key is accessed or updated, move it to the list for `freq + 1` and update `min_freq` if necessary.

Step 5: During a `put` operation, if the cache is full, remove the tail node from the `min_freq` list.

```python
import collections

class Node:
    def __init__(self, key, val):
        self.key = key
        self.val = val
        self.freq = 1
        self.prev = self.next = None

class DLinkedList:
    def __init__(self):
        self.sentinel = Node(None, None)
        self.sentinel.next = self.sentinel.prev = self.sentinel
        self.size = 0
    def append(self, node):
        node.next = self.sentinel.next
        node.prev = self.sentinel
        self.sentinel.next.prev = node
        self.sentinel.next = node
        self.size += 1
    def pop(self, node=None):
        if self.size == 0: return None
        if not node: node = self.sentinel.prev
        node.prev.next = node.next
        node.next.prev = node.prev
        self.size -= 1
        return node

class LFUCache:
    def __init__(self, capacity: int):
        self.cap = capacity
        self.nodes = {}
        self.freqs = collections.defaultdict(DLinkedList)
        self.min_freq = 0

    def get(self, key: int) -> int:
        if key not in self.nodes: return -1
        node = self.nodes[key]
        self._update(node)
        return node.val

    def put(self, key: int, value: int) -> None:
        if self.cap == 0: return
        if key in self.nodes:
            node = self.nodes[key]
            node.val = value
            self._update(node)
        else:
            if len(self.nodes) == self.cap:
                old_node = self.freqs[self.min_freq].pop()
                del self.nodes[old_node.key]
            new_node = Node(key, value)
            self.nodes[key] = new_node
            self.freqs[1].append(new_node)
            self.min_freq = 1

    def _update(self, node):
        freq = node.freq
        self.freqs[freq].pop(node)
        if self.min_freq == freq and self.freqs[freq].size == 0:
            self.min_freq += 1
        node.freq += 1
        self.freqs[node.freq].append(node)
```

### [Linked List] Problem 148: Sort List

**Algorithm & Technique:** Merge Sort

**Logic:**
Step 1: Use the slow and fast pointer technique to find the middle of the linked list.

Step 2: Split the list into two halves and recursively call the sort function on both halves.

Step 3: Implement a `merge` helper function to combine two sorted linked lists into one.

Step 4: Base case: If the list is empty or has only one node, it is already sorted.

Step 5: Return the merged sorted list.

```python
class Solution:
    def sortList(self, head: ListNode) -> ListNode:
        if not head or not head.next: return head
        fast, slow = head.next, head
        while fast and fast.next:
            fast = fast.next.next
            slow = slow.next
        mid = slow.next
        slow.next = None
        left = self.sortList(head)
        right = self.sortList(mid)
        return self.merge(left, right)

    def merge(self, l1, l2):
        dummy = curr = ListNode(0)
        while l1 and l2:
            if l1.val < l2.val:
                curr.next, l1 = l1, l1.next
            else:
                curr.next, l2 = l2, l2.next
            curr = curr.next
        curr.next = l1 or l2
        return dummy.next
```

### [Searching] Problem 4: Median of Two Sorted Arrays

**Algorithm & Technique:** Binary Search on Partitions

**Logic:**
Step 1: Ensure we perform binary search on the smaller of the two arrays to optimize time complexity to $O(\log(\min(m, n)))$.

Step 2: Partition both arrays such that the total number of elements on the left side is half the total size.

Step 3: Check if the maximum element on the left is less than or equal to the minimum element on the right for both arrays.

Step 4: If conditions are met, calculate the median based on whether the total number of elements is even or odd.

Step 5: If the partition is incorrect, adjust the binary search range left or right.

```python
class Solution:
    def findMedianSortedArrays(self, nums1: list[int], nums2: list[int]) -> float:
        if len(nums1) > len(nums2): nums1, nums2 = nums2, nums1
        m, n = len(nums1), len(nums2)
        low, high = 0, m
        while low <= high:
            partitionX = (low + high) // 2
            partitionY = (m + n + 1) // 2 - partitionX
            maxLeftX = float('-inf') if partitionX == 0 else nums1[partitionX - 1]
            minRightX = float('inf') if partitionX == m else nums1[partitionX]
            maxLeftY = float('-inf') if partitionY == 0 else nums2[partitionY - 1]
            minRightY = float('inf') if partitionY == n else nums2[partitionY]
            if maxLeftX <= minRightY and maxLeftY <= minRightX:
                if (m + n) % 2 == 0:
                    return (max(maxLeftX, maxLeftY) + min(minRightX, minRightY)) / 2
                else:
                    return max(maxLeftX, maxLeftY)
            elif maxLeftX > minRightY: high = partitionX - 1
            else: low = partitionX + 1
```

### [Searching] Problem 154: Find Minimum in Rotated Sorted Array II

**Algorithm & Technique:** Binary Search

**Logic:**
Step 1: Initialize `low` and `high` pointers at the ends of the array.

Step 2: Calculate the `mid` index.

Step 3: If `nums[mid] > nums[high]`, the minimum must be in the right half; set `low = mid + 1`.

Step 4: If `nums[mid] < nums[high]`, the minimum is in the left half or at `mid`; set `high = mid`.

Step 5: If `nums[mid] == nums[high]`, we cannot be sure; decrement `high` to narrow the search.

Step 6: Return `nums[low]` once the pointers converge.

```python
class Solution:
    def findMin(self, nums: list[int]) -> int:
        low, high = 0, len(nums) - 1
        while low < high:
            mid = (low + high) // 2
            if nums[mid] > nums[high]: low = mid + 1
            elif nums[mid] < nums[high]: high = mid
            else: high -= 1
        return nums[low]
```

### [Searching] Problem 410: Split Array Largest Sum

**Algorithm & Technique:** Binary Search on Answer

**Logic:**
Step 1: Define the search range for the answer: `low` is the max element and `high` is the sum of all elements.

Step 2: Use binary search to pick a `mid` value as a potential maximum subarray sum.

Step 3: Use a greedy helper function to check if the array can be split into $m$ or fewer subarrays with a sum not exceeding `mid`.

Step 4: If possible, try a smaller sum by setting `high = mid`.

Step 5: If not possible, increase the allowed sum by setting `low = mid + 1`.

Step 6: Return `low` as the minimized largest sum.

```python
class Solution:
    def splitArray(self, nums: list[int], m: int) -> int:
        def canSplit(max_sum):
            count, curr_sum = 1, 0
            for num in nums:
                curr_sum += num
                if curr_sum > max_sum:
                    count += 1
                    curr_sum = num
            return count <= m
        low, high = max(nums), sum(nums)
        while low < high:
            mid = (low + high) // 2
            if canSplit(mid): high = mid
            else: low = mid + 1
        return low
```

### [Searching] Problem 378: Kth Smallest Element in a Sorted Matrix

**Algorithm & Technique:** Binary Search on Value Range

**Logic:**
Step 1: Initialize search range with the smallest value (top-left) and largest value (bottom-right) of the matrix.

Step 2: Pick a `mid` value and count how many elements in the matrix are less than or equal to it.

Step 3: Use the property that rows and columns are sorted to count in $O(n)$ time.

Step 4: If the count is less than $k$, move the range to the right half (`low = mid + 1`).

Step 5: Otherwise, move to the left half (`high = mid`).

Step 6: Return `low` as the $k$-th smallest element.

```python
class Solution:
    def kthSmallest(self, matrix: list[list[int]], k: int) -> int:
        n = len(matrix)
        low, high = matrix[0][0], matrix[-1][-1]
        while low < high:
            mid = (low + high) // 2
            count = 0
            j = n - 1
            for i in range(n):
                while j >= 0 and matrix[i][j] > mid: j -= 1
                count += (j + 1)
            if count < k: low = mid + 1
            else: high = mid
        return low
```

### [Sorting] Problem 56: Merge Intervals

**Algorithm & Technique:** Sorting + Greedy

**Logic:**

Step 3: Iterate through each interval.

### [Sorting] Problem 315: Count of Smaller Numbers After Self

**Algorithm & Technique:** Merge Sort based Counting

**Logic:**
Step 1: Use an array of indices to track positions and avoid moving original values.

Step 2: Perform a standard merge sort on the indices based on the values in the original array.

Step 3: During the merge step, when an element from the left half is merged, count how many elements from the right half have already been merged.

Step 4: This count represents the number of elements to the right that are smaller than the current element.

Step 5: Store and accumulate these counts in a result array.

Step 6: Return the accumulated counts.

```python
class Solution:
    def countSmaller(self, nums: list[int]) -> list[int]:
        n = len(nums)
        counts = [0] * n
        indices = list(range(n))
        def merge_sort(left, right):
            if right - left <= 1: return
            mid = (left + right) // 2
            merge_sort(left, mid)
            merge_sort(mid, right)
            merge(left, mid, right)
        def merge(left, mid, right):
            i, j = left, mid
            temp = []
            while i < mid and j < right:
                if nums[indices[i]] <= nums[indices[j]]:
                    counts[indices[i]] += j - mid
                    temp.append(indices[i])
                    i += 1
                else:
                    temp.append(indices[j])
                    j += 1
            while i < mid:
                counts[indices[i]] += j - mid
                temp.append(indices[i])
                i += 1
            while j < right:
                temp.append(indices[j])
                j += 1
            indices[left:right] = temp
        merge_sort(0, n)
        return counts
```

### [Sorting] Problem 354: Russian Doll Envelopes

**Algorithm & Technique:** Sorting + Longest Increasing Subsequence (LIS)

**Logic:**
Step 1: Sort the envelopes: ascending by width, and descending by height for same widths.

Step 2: Extract the heights into a new array.

Step 3: Find the Longest Increasing Subsequence (LIS) of the heights array.

Step 4: Use binary search (patience sorting) to find the LIS in $O(n \log n)$ time.

Step 5: The length of the LIS is the maximum number of envelopes that can be nested.

```python
import bisect
class Solution:
    def maxEnvelopes(self, envelopes: list[list[int]]) -> int:
        envelopes.sort(key=lambda x: (x[0], -x[1]))
        heights = [e[1] for e in envelopes]
        dp = []
        for h in heights:
            idx = bisect.bisect_left(dp, h)
            if idx == len(dp):
                dp.append(h)
            else:
                dp[idx] = h
        return len(dp)
```

### [Sorting] Problem 164: Maximum Gap

**Algorithm & Technique:**
This problem is solved using **Bucket Sort** based on the **Pigeonhole Principle**. Since we need to find the maximum gap in linear time $O(n)$, we cannot use standard comparison-based sorting ($O(n \log n)$).

**Logic:**
1.  **Bucket Range:** We calculate the minimum and maximum values in the array. We then divide the range into $n-1$ buckets of equal size.
2.  **Pigeonhole Principle:** By creating $n-1$ buckets for $n$ numbers, at least one bucket (or the gap between buckets) must contain the maximum gap, and the maximum gap cannot occur within a single bucket if the bucket size is smaller than the average gap.
3.  **Tracking:** For each bucket, we only store the `min` and `max` values because the maximum gap will always occur between the `max` of one non-empty bucket and the `min` of the next non-empty bucket.
4.  **Gap Calculation:** We iterate through the sorted buckets to find the largest difference between a current bucket's minimum and the previous bucket's maximum.

```python
class Solution:
    def maximumGap(self, nums: list[int]) -> int:
        if len(nums) < 2: return 0
        min_val, max_val = min(nums), max(nums)
        if min_val == max_val: return 0
        n = len(nums)
        bucket_size = max(1, (max_val - min_val) // (n - 1))
        bucket_count = (max_val - min_val) // bucket_size + 1
        buckets = [[None, None] for _ in range(bucket_count)]
        for num in nums:
            idx = (num - min_val) // bucket_size
            if buckets[idx][0] is None:
                buckets[idx][0] = buckets[idx][1] = num
            else:
                buckets[idx][0] = min(buckets[idx][0], num)
                buckets[idx][1] = max(buckets[idx][1], num)
        max_gap = 0
        prev_max = min_val
        for i in range(bucket_count):
            if buckets[i][0] is None: continue
            max_gap = max(max_gap, buckets[i][0] - prev_max)
            prev_max = buckets[i][1]
        return max_gap
```

### [Trees] Problem 124: Binary Tree Maximum Path Sum

**Algorithm & Technique:** Recursion (DFS)

**Logic:**
Step 1: Define a recursive function that returns the maximum path sum starting from a node and going downwards.

Step 2: For each node, calculate the max contribution from its left and right children (ignoring negative sums).

Step 3: Update a global maximum variable with the sum of the current node's value plus the left and right contributions.

Step 4: Return the node's value plus the better of the two children's contributions to the parent.

Step 5: After DFS, the global variable holds the result.

```python
class Solution:
    def maxPathSum(self, root: TreeNode) -> int:
        self.res = float('-inf')
        def dfs(node):
            if not node: return 0
            left = max(0, dfs(node.left))
            right = max(0, dfs(node.right))
            self.res = max(self.res, left + right + node.val)
            return node.val + max(left, right)
        dfs(root)
        return self.res
```

### [Trees] Problem 99: Recover Binary Search Tree

**Logic:**
Step 1: Perform an in-order traversal of the tree, which should result in sorted values for a BST.

Step 2: Keep track of the previously visited node (`prev`).

Step 3: Identify the two nodes where the sorted order is violated (`prev.val > curr.val`).

Step 4: The first swapped node is the `prev` of the first violation; the second is the `curr` of the last violation.

Step 5: Swap the values of these two identified nodes to restore the BST property.

```python
class Solution:
    def recoverTree(self, root: TreeNode) -> None:
        self.first = self.second = self.prev = None
        def inorder(node):
            if not node: return
            inorder(node.left)
            if self.prev and self.prev.val > node.val:
                if not self.first: self.first = self.prev
                self.second = node
            self.prev = node
            inorder(node.right)
        inorder(root)
        self.first.val, self.second.val = self.second.val, self.first.val
```

**Algorithm & Technique:** Binary Search / Recursion

**Logic:**
Step 1: Calculate the height of the left-most path and the right-most path.

Step 2: If the two heights are equal, the tree is a perfect binary tree, and the number of nodes is $2^h - 1$.

Step 3: If the heights are not equal, the total count is $1 + \text{countNodes(left)} + \text{countNodes(right)}$.

Step 4: This approach ensures we don't visit every node, keeping complexity better than $O(n)$.

```python
class Solution:
    def countNodes(self, root: TreeNode) -> int:
        if not root: return 0
        l, r = root, root
        hl, hr = 0, 0
        while l: hl += 1; l = l.left
        while r: hr += 1; r = r.right
        if hl == hr: return (2**hl) - 1
        return 1 + self.countNodes(root.left) + self.countNodes(root.right)
```

### [Stack] Problem 772: Basic Calculator III

**Algorithm & Technique:** Two Stacks / Recursion

**Logic:**
Step 1: Create a function that parses the string and uses a stack to store numbers and intermediate results.

Step 2: Use a variable `num` to build digits and `sign` to track the last operator (+, -, *, /).

Step 3: When encountering a '(', recursively call the function to evaluate the sub-expression.

Step 4: When encountering a ')', return the current result to the parent caller.

Step 5: Process operators: for '*' and '/', pop from stack, perform operation with `num`, and push back. For '+' and '-', push `num` or `-num`.

Step 6: Sum the stack at the end to get the final result.

```python
class Solution:
    def calculate(self, s: str) -> int:
        s += '+'
        stack = []
        num = 0
        sign = '+'
        i = 0
        while i < len(s):
            char = s[i]
            if char.isdigit():
                num = num * 10 + int(char)
            elif char == '(':
                # Find matching parenthesis to recurse
                start = i
                count = 0
                while i < len(s):
                    if s[i] == '(': count += 1
                    elif s[i] == ')': count -= 1
                    if count == 0: break
                    i += 1
                num = self.calculate(s[start + 1:i])
            elif char in '+-*/':
                if sign == '+': stack.append(num)
                elif sign == '-': stack.append(-num)
                elif sign == '*': stack.append(stack.pop() * num)
                elif sign == '/': stack.append(int(stack.pop() / num))
                sign = char
                num = 0
            i += 1
        return sum(stack)
```

**Logic:**
Step 1: Treat each row of the matrix as the base of a histogram.

Step 2: Maintain an array `heights` where `heights[j]` is the number of consecutive 1s ending at current row $i$, column $j$.

Step 3: If `matrix[i][j]` is '0', reset `heights[j]` to 0.

Step 4: For every row, use the 'Largest Rectangle in Histogram' (Problem 84) algorithm to find the max area for that row's histogram.

Step 5: Update the global maximum area across all rows.

```python
class Solution:
    def maximalRectangle(self, matrix: list[list[str]]) -> int:
        if not matrix: return 0
        n = len(matrix[0])
        heights = [0] * (n + 1)
        max_area = 0
        for row in matrix:
            for i in range(n):
                heights[i] = heights[i] + 1 if row[i] == '1' else 0

            stack = [-1]
            for i in range(n + 1):
                while heights[i] < heights[stack[-1]]:
                    h = heights[stack.pop()]
                    w = i - stack[-1] - 1
                    max_area = max(max_area, h * w)
                stack.append(i)
        return max_area
```

### [Queue] Problem 239: Sliding Window Maximum

**Algorithm & Technique:** Monotonic Deque

**Logic:**
Step 1: Create a deque to store indices of elements in the current window.

Step 2: Iterate through the array.

Step 3: Remove indices from the front of the deque if they are outside the current window range.

Step 4: Remove indices from the back of the deque if their corresponding values are smaller than the current element.

Step 5: Append the current index to the deque.

Step 6: The front of the deque always points to the maximum value in the current window.

```python
from collections import deque
class Solution:
    def maxSlidingWindow(self, nums: list[int], k: int) -> list[int]:
        q = deque()
        res = []
        for i, n in enumerate(nums):
            while q and nums[q[-1]] <= n:
                q.pop()
            q.append(i)
            if q[0] == i - k:
                q.popleft()
            if i >= k - 1:
                res.append(nums[q[0]])
        return res
```

### [Graphs] Problem 210: Course Schedule II

**Algorithm & Technique:** Topological Sort (Kahn's Algorithm)

**Logic:**
Step 1: Calculate the in-degree of every node and build an adjacency list.

Step 2: Add all nodes with an in-degree of 0 to a queue.

Step 3: Pop a node from the queue and add it to the topological order result.

Step 4: For each neighbor of the popped node, decrement its in-degree.

Step 5: If a neighbor's in-degree becomes 0, add it to the queue.

Step 6: If the result length equals the number of nodes, return the order; otherwise, a cycle exists.

```python
from collections import deque, defaultdict
class Solution:
    def findOrder(self, numCourses: int, prerequisites: list[list[int]]) -> list[int]:
        adj = defaultdict(list)
        indegree = [0] * numCourses
        for dest, src in prerequisites:
            adj[src].append(dest)
            indegree[dest] += 1

        q = deque([i for i in range(numCourses) if indegree[i] == 0])
        res = []
        while q:
            u = q.popleft()
            res.append(u)
            for v in adj[u]:
                indegree[v] -= 1
                if indegree[v] == 0:
                    q.append(v)
        return res if len(res) == numCourses else []
```

### [Graphs] Problem 126: Word Ladder II

**Algorithm & Technique:** BFS + Backtracking (DFS)

**Logic:**
Step 1: Use BFS to find the shortest distance from the `beginWord` to all reachable words.

Step 2: During BFS, build an adjacency map to store the parents of each word that lead to the shortest path.

Step 3: Once the `endWord` is reached, stop the BFS at that level.

Step 4: Use DFS (backtracking) starting from the `endWord` to reconstruct all shortest paths using the parent map.

Step 5: Reverse the reconstructed paths and return them.

```python
from collections import defaultdict, deque
class Solution:
    def findLadders(self, beginWord: str, endWord: str, wordList: list[str]) -> list[list[str]]:
        wordSet = set(wordList)
        if endWord not in wordSet: return []
        layer = {beginWord: [[beginWord]]}
        while layer:
            new_layer = defaultdict(list)
            for word in layer:
                if word == endWord: return layer[word]
                for i in range(len(word)):
                    for c in 'abcdefghijklmnopqrstuvwxyz':
                        next_word = word[:i] + c + word[i+1:]
                        if next_word in wordSet:
                            for path in layer[word]:
                                new_layer[next_word].append(path + [next_word])
            wordSet -= set(new_layer.keys())
            layer = new_layer
        return []
```

### [Graphs] Problem 269: Alien Dictionary

**Algorithm & Technique:** Topological Sort (BFS/Kahn's)

**Logic:**
Step 1: Build a graph where an edge $A \to B$ exists if character $A$ comes before $B$ in the dictionary.

Step 2: Compare adjacent words in the list to find the first differing character to create edges.

Step 3: Handle edge case where a word is a prefix of the previous word (invalid order).

Step 4: Perform topological sort on the character graph.

Step 5: If the topological order contains all unique characters from the input, return the string; else return "".

```python
from collections import deque, defaultdict
class Solution:
    def alienOrder(self, words: list[str]) -> str:
        adj = {c: set() for w in words for c in w}
        indegree = {c: 0 for w in words for c in w}
        for i in range(len(words) - 1):
            w1, w2 = words[i], words[i+1]
            minLen = min(len(w1), len(w2))
            if len(w1) > len(w2) and w1[:minLen] == w2[:minLen]: return ""
            for j in range(minLen):
                if w1[j] != w2[j]:
                    if w2[j] not in adj[w1[j]]:
                        adj[w1[j]].add(w2[j])
                        indegree[w2[j]] += 1
                    break
        q = deque([c for c in indegree if indegree[c] == 0])
        res = []
        while q:
            c = q.popleft()
            res.append(c)
            for neighbor in adj[c]:
                indegree[neighbor] -= 1
                if indegree[neighbor] == 0: q.append(neighbor)
        return "".join(res) if len(res) == len(indegree) else ""
```

### [Graphs] Problem 1192: Critical Connections in a Network

**Algorithm & Technique:** Tarjan's Algorithm (Bridges in Graph)

**Logic:**
Step 1: Use DFS to traverse the graph and assign a discovery time and low-link value to each node.

Step 2: A critical connection (bridge) is an edge $(u, v)$ where the low-link value of $v$ is strictly greater than the discovery time of $u$.

Step 3: During DFS, update the low-link value of $u$ based on its neighbors' low-link values (excluding its parent).

Step 4: If $low[v] > disc[u]$, add $[u, v]$ to the result list.

```python
from collections import defaultdict
class Solution:
    def criticalConnections(self, n: int, connections: list[list[int]]) -> list[list[int]]:
        adj = defaultdict(list)
        for u, v in connections:
            adj[u].append(v)
            adj[v].append(u)
        disc, low, res = [-1] * n, [-1] * n, []
        def dfs(u, p, time):
            disc[u] = low[u] = time
            for v in adj[u]:
                if v == p: continue
                if disc[v] == -1:
                    dfs(v, u, time + 1)
                    low[u] = min(low[u], low[v])
                    if low[v] > disc[u]: res.append([u, v])
                else:
                    low[u] = min(low[u], disc[v])
        dfs(0, -1, 0)
        return res
```

### [Graphs] Problem 847: Shortest Path Visiting All Nodes

**Algorithm & Technique:** BFS + Bitmask

**Logic:**
Step 1: Use BFS to find the shortest path, where the state is defined as `(current_node, visited_mask)`.

Step 2: The `visited_mask` is an integer where the $i$-th bit is set if node $i$ has been visited.

Step 3: Initialize the queue with all nodes as starting points: `(i, 1 << i)` with distance 0.

Step 4: In each step, move to an adjacent node and update the mask.

Step 5: If the mask equals $(1 \ll n) - 1$, all nodes have been visited; return the distance.

```python
from collections import deque
class Solution:
    def shortestPathLength(self, graph: list[list[int]]) -> int:
        n = len(graph)
        target = (1 << n) - 1
        q = deque([(i, 1 << i, 0) for i in range(n)])
        visited = set([(i, 1 << i) for i in range(n)])
        while q:
            u, mask, dist = q.popleft()
            if mask == target: return dist
            for v in graph[u]:
                new_mask = mask | (1 << v)
                if (v, new_mask) not in visited:
                    visited.add((v, new_mask))
                    q.append((v, new_mask, dist + 1))
```

### [Hashing] Problem 128: Longest Consecutive Sequence

**Algorithm & Technique:** Hash Set

**Logic:**
Step 1: Insert all numbers into a hash set for $O(1)$ lookups.

Step 2: Iterate through the set elements.

Step 3: For each element `n`, check if `n - 1` is in the set. If it isn't, `n` is the start of a sequence.

Step 4: If `n` is a start, count how many consecutive numbers follow (`n+1`, `n+2`, etc.).

Step 5: Update the maximum length found so far.

```python
class Solution:
    def longestConsecutive(self, nums: list[int]) -> int:
        numSet = set(nums)
        maxLen = 0
        for n in numSet:
            if n - 1 not in numSet:
                curr = n
                currLen = 1
                while curr + 1 in numSet:
                    curr += 1
                    currLen += 1
                maxLen = max(maxLen, currLen)
        return maxLen
```

### [Hashing] Problem 560: Subarray Sum Equals K

**Algorithm & Technique:** Prefix Sum + Hash Map

**Logic:**
Step 1: Use a hash map to store the frequency of prefix sums encountered so far.

Step 2: Initialize the map with `{0: 1}` to handle cases where the subarray starts from index 0.

Step 3: Iterate through the array and calculate the cumulative prefix sum.

Step 4: If `prefix_sum - k` exists in the map, add its frequency to the total count.

Step 5: Update the frequency of the current prefix sum in the map.

```python
from collections import defaultdict
class Solution:
    def subarraySum(self, nums: list[int], k: int) -> int:
        count = 0
        curr_sum = 0
        sums = defaultdict(int)
        sums[0] = 1
        for n in nums:
            curr_sum += n
            count += sums[curr_sum - k]
            sums[curr_sum] += 1
        return count
```

### [Hashing] Problem 325: Maximum Size Subarray Sum Equals k

**Algorithm & Technique:** Prefix Sum + Hash Map

**Logic:**
Step 1: Initialize a hash map to store the first occurrence of each prefix sum.

Step 2: Store `{0: -1}` to handle cases where the subarray starts from index 0.

Step 3: Iterate through the array and maintain a running prefix sum.

Step 4: If `prefix_sum - k` is in the map, calculate the length of the subarray ending at the current index.

Step 5: Update the maximum length found so far.

Step 6: If the current prefix sum is not in the map, store it with its index.

```python
class Solution:
    def maxSubArrayLen(self, nums: list[int], k: int) -> int:
        prefix_map = {0: -1}
        curr_sum = 0
        max_len = 0
        for i, num in enumerate(nums):
            curr_sum += num
            if curr_sum - k in prefix_map:
                max_len = max(max_len, i - prefix_map[curr_sum - k])
            if curr_sum not in prefix_map:
                prefix_map[curr_sum] = i
        return max_len
```

### [Hashing] Problem 76: Minimum Window Substring

**Algorithm & Technique:** Sliding Window + Frequency Map

**Logic:**
Step 1: Count frequencies of characters in string `t`.

Step 2: Use two pointers (`left`, `right`) to expand a window over string `s`.

Step 3: Maintain a count of characters currently matched within the window.

Step 4: Once all characters of `t` are covered, shrink the window from the left to find the minimum length.

Step 5: Record the start index and minimum length of the smallest valid window found.

```python
from collections import Counter
class Solution:
    def minWindow(self, s: str, t: str) -> str:
        if not t or not s: return ""
        dict_t = Counter(t)
        required = len(dict_t)
        l, r = 0, 0
        formed = 0
        window_counts = {}
        ans = float("inf"), None, None
        while r < len(s):
            char = s[r]
            window_counts[char] = window_counts.get(char, 0) + 1
            if char in dict_t and window_counts[char] == dict_t[char]:
                formed += 1
            while l <= r and formed == required:
                char = s[l]
                if r - l + 1 < ans[0]:
                    ans = (r - l + 1, l, r)
                window_counts[char] -= 1
                if char in dict_t and window_counts[char] < dict_t[char]:
                    formed -= 1
                l += 1
            r += 1
        return "" if ans[0] == float("inf") else s[ans[1] : ans[2] + 1]
```

### [Recursion (Backtracking)] Problem 37: Sudoku Solver

**Algorithm & Technique:** Backtracking

**Logic:**
Step 1: Find the next empty cell on the board.

Step 2: Try placing digits '1' through '9' in that cell.

Step 3: For each digit, check if it's valid according to Sudoku rules (row, column, 3x3 box).

Step 4: If valid, place the digit and recursively attempt to solve the rest of the board.

Step 5: If the recursive call returns False, backtrack by resetting the cell and trying the next digit.

Step 6: Return True if the board is completely and correctly filled.

```python
class Solution:
    def solveSudoku(self, board: list[list[str]]) -> None:
        def isValid(r, c, char):
            for i in range(9):
                if board[i][c] == char: return False
                if board[r][i] == char: return False
                if board[3 * (r // 3) + i // 3][3 * (c // 3) + i % 3] == char: return False
            return True
        def solve():
            for i in range(9):
                for j in range(9):
                    if board[i][j] == '.':
                        for char in '123456789':
                            if isValid(i, j, char):
                                board[i][j] = char
                                if solve(): return True
                                board[i][j] = '.'
                        return False
            return True
        solve()
```

### [Recursion (Backtracking)] Problem 282: Expression Add Operators

**Algorithm & Technique:** Backtracking with Evaluation

**Logic:**
Step 1: Use a recursive function to build expressions by picking substrings and operators (+, -, *).

Step 2: Track the current calculated value and the 'previous operand' to handle operator precedence for multiplication.

Step 3: Base case: if the end of the string is reached and the value equals the target, add to the result.

Step 4: Avoid numbers with leading zeros.

Step 5: For multiplication, the logic is: `prev_val - last_op + (last_op * curr_num)`.

```python
class Solution:
    def addOperators(self, num: str, target: int) -> list[str]:
        res = []
        def backtrack(idx, path, value, last_op):
            if idx == len(num):
                if value == target: res.append(path)
                return
            for i in range(idx, len(num)):
                if i > idx and num[idx] == '0': break
                s = num[idx:i+1]
                n = int(s)
                if idx == 0:
                    backtrack(i + 1, s, n, n)
                else:
                    backtrack(i + 1, path + '+' + s, value + n, n)
                    backtrack(i + 1, path + '-' + s, value - n, -n)
                    backtrack(i + 1, path + '*' + s, value - last_op + last_op * n, last_op * n)
        backtrack(0, "", 0, 0)
        return res
```

### [Recursion (Backtracking)] Problem 131: Palindrome Partitioning

**Algorithm & Technique:** Backtracking

**Logic:**
Step 1: Use a recursive function to split the string into prefixes.

Step 2: For each split, check if the current prefix is a palindrome.

Step 3: If it is, add it to the current path and recursively process the remaining suffix.

Step 4: If the end of the string is reached, add the current path to the result list.

Step 5: Backtrack by removing the prefix from the path after the recursive call.

```python
class Solution:
    def partition(self, s: str) -> list[list[str]]:
        res = []
        def backtrack(start, path):
            if start == len(s): res.append(path[:])
            for end in range(start + 1, len(s) + 1):
                if s[start:end] == s[start:end][::-1]:
                    path.append(s[start:end])
                    backtrack(end, path)
                    path.pop()
        backtrack(0, [])
        return res
```

### [Recursion (Backtracking)] Problem 212: Word Search II

**Algorithm & Technique:** Backtracking + Trie

**Logic:**
Step 1: Build a Trie (Prefix Tree) containing all words in the dictionary.

Step 2: Traverse each cell of the grid and start a DFS if the starting character is in the Trie.

Step 3: During DFS, mark cells as visited and move in four directions.

Step 4: If a word is found at the current Trie node, add it to the result and remove it from the Trie to avoid duplicates.

Step 5: Backtrack by restoring the cell's original character after the search.

```python
class Solution:
    def findWords(self, board: list[list[str]], words: list[str]) -> list[str]:
        trie = {}
        for w in words:
            node = trie
            for char in w:
                node = node.setdefault(char, {})
            node['#'] = w
        res, rows, cols = set(), len(board), len(board[0])
        def dfs(r, c, node):
            char = board[r][c]
            curr_node = node[char]
            word = curr_node.pop('#', None)
            if word: res.add(word)
            board[r][c] = '*'
            for dr, dc in [(0,1),(0,-1),(1,0),(-1,0)]:
                nr, nc = r+dr, c+dc
                if 0<=nr<rows and 0<=nc<cols and board[nr][nc] in curr_node:
                    dfs(nr, nc, curr_node)
            board[r][c] = char
            if not curr_node: node.pop(char)
        for r in range(rows):
            for c in range(cols):
                if board[r][c] in trie: dfs(r, c, trie)
        return list(res)
```

### [Dynamic Programming] Problem 72: Edit Distance

**Algorithm & Technique:** 2D Dynamic Programming

**Logic:**
Step 1: Create a 2D table `dp` where `dp[i][j]` is the edit distance between `word1[:i]` and `word2[:j]`.

Step 2: Initialize base cases: `dp[i][0] = i` (deletions) and `dp[0][j] = j` (insertions).

Step 3: If `word1[i-1] == word2[j-1]`, then `dp[i][j] = dp[i-1][j-1]`.

Step 4: Else, `dp[i][j] = 1 + min(dp[i-1][j], dp[i][j-1], dp[i-1][j-1])` (representing Delete, Insert, and Replace operations).

Step 5: Return the value at `dp[m][n]`.

```python
class Solution:
    def minDistance(self, word1: str, word2: str) -> int:
        m, n = len(word1), len(word2)
        dp = [[0] * (n + 1) for _ in range(m + 1)]
        for i in range(m + 1): dp[i][0] = i
        for j in range(n + 1): dp[0][j] = j
        for i in range(1, m + 1):
            for j in range(1, n + 1):
                if word1[i-1] == word2[j-1]:
                    dp[i][j] = dp[i-1][j-1]
                else:
                    dp[i][j] = 1 + min(dp[i-1][j], dp[i][j-1], dp[i-1][j-1])
        return dp[m][n]
```

### [Dynamic Programming] Problem 312: Burst Balloons

**Algorithm & Technique:** Interval Dynamic Programming

**Logic:**
Step 1: Add boundary balloons with value 1 at both ends of the input array.

Step 2: Use `dp[i][j]` to represent the maximum coins obtained by bursting all balloons in the open interval $(i, j)$.

Step 3: Iterate through all possible interval lengths from 1 to $n$.

Step 4: For each interval $(i, j)$, pick a balloon $k$ to be the **last** one to burst in that interval.

Step 5: Calculate coins: `nums[i] * nums[k] * nums[j] + dp[i][k] + dp[k][j]`.

Step 6: Update `dp[i][j]` with the maximum value found.

```python
class Solution:
    def maxCoins(self, nums: list[int]) -> int:
        A = [1] + [x for x in nums if x > 0] + [1]
        n = len(A)
        dp = [[0] * n for _ in range(n)]
        for length in range(2, n):
            for i in range(n - length):
                j = i + length
                for k in range(i + 1, j):
                    dp[i][j] = max(dp[i][j], A[i]*A[k]*A[j] + dp[i][k] + dp[k][j])
        return dp[0][n-1]
```

### [Dynamic Programming] Problem 10: Regular Expression Matching

**Algorithm & Technique:** 2D Dynamic Programming / Memoization

**Logic:**
Step 1: Use `dp[i][j]` to represent if `s[i:]` matches `p[j:]`.

Step 2: Handle the '*' character which matches zero or more of the preceding element.

Step 3: If `p[j+1]` is '*', we can either skip the pattern (`dp(i, j+2)`) or use the character if it matches `s[i]` (`match and dp(i+1, j)`).

Step 4: If no '*', check if the current character matches and recurse to `dp(i+1, j+1)`.

Step 5: Return the result for `dp(0, 0)`.

```python
class Solution:
    def isMatch(self, s: str, p: str) -> bool:
        memo = {}
        def dp(i, j):
            if (i, j) in memo: return memo[(i, j)]
            if j == len(p): return i == len(s)
            match = i < len(s) and (s[i] == p[j] or p[j] == '.')
            if j + 1 < len(p) and p[j+1] == '*':
                res = dp(i, j + 2) or (match and dp(i + 1, j))
            else:
                res = match and dp(i + 1, j + 1)
            memo[(i, j)] = res
            return res
        return dp(0, 0)
```

### [Dynamic Programming] Problem 115: Distinct Subsequences

**Algorithm & Technique:** 2D Dynamic Programming

**Logic:**
Step 1: Use `dp[i][j]` to represent the number of times `t[:i]` appears as a subsequence in `s[:j]`.

Step 2: Initialize `dp[0][j] = 1` for all $j$ (an empty target is found once in any string).

Step 3: If `t[i-1] == s[j-1]`, the target can be formed by including the character (`dp[i-1][j-1]`) or excluding it (`dp[i][j-1]`).

Step 4: If characters don't match, `dp[i][j] = dp[i][j-1]`.

Step 5: Return `dp[m][n]`.

```python
class Solution:
    def numDistinct(self, s: str, t: str) -> int:
        m, n = len(t), len(s)
        dp = [[0] * (n + 1) for _ in range(m + 1)]
        for j in range(n + 1): dp[0][j] = 1
        for i in range(1, m + 1):
            for j in range(1, n + 1):
                if t[i-1] == s[j-1]:
                    dp[i][j] = dp[i-1][j-1] + dp[i][j-1]
                else:
                    dp[i][j] = dp[i][j-1]
        return dp[m][n]
```

### [Dynamic Programming] Problem 132: Palindrome Partitioning II

**Algorithm & Technique:** Dynamic Programming + Precomputed Palindromes

**Logic:**
Step 1: Precompute a 2D boolean table `isPal[i][j]` to check if `s[i:j+1]` is a palindrome.

Step 2: Create a `cuts` array where `cuts[i]` is the minimum cuts needed for `s[:i]`.

Step 3: Initialize `cuts[i] = i - 1` (maximum possible cuts).

Step 4: For each `j`, if `s[0:j+1]` is a palindrome, `cuts[j] = 0`.

Step 5: Else, check all splits $i < j$; if `s[i:j+1]` is a palindrome, `cuts[j] = min(cuts[j], cuts[i-1] + 1)`.

Step 6: Return `cuts[n-1]`.

```python
class Solution:
    def minCut(self, s: str) -> int:
        n = len(s)
        isPal = [[False] * n for _ in range(n)]
        for length in range(n):
            for i in range(n - length):
                j = i + length
                if s[i] == s[j] and (j - i < 2 or isPal[i+1][j-1]):
                    isPal[i][j] = True
        cuts = [i for i in range(n)]
        for j in range(n):
            if isPal[0][j]:
                cuts[j] = 0
            else:
                for i in range(1, j + 1):
                    if isPal[i][j]:
                        cuts[j] = min(cuts[j], cuts[i-1] + 1)
        return cuts[n-1]
```

**Algorithm & Technique:** Recursion / Hash Map

Step 5: Recursively build the left subtree using the range to the left of the root's inorder index.

Step 6: Recursively build the right subtree using the range to the right of the root's inorder index.

```python
class Solution:
    def buildTree(self, preorder: list[int], inorder: list[int]) -> TreeNode:
        inorder_map = {val: i for i, val in enumerate(inorder)}
        self.pre_idx = 0
        def helper(in_left, in_right):
            if in_left > in_right: return None
            root_val = preorder[self.pre_idx]
            root = TreeNode(root_val)
            self.pre_idx += 1
            idx = inorder_map[root_val]
            root.left = helper(in_left, idx - 1)
            root.right = helper(idx + 1, in_right)
            return root
        return helper(0, len(inorder) - 1)
```

### [Trees] Problem 437: Path Sum III

**Algorithm & Technique:** Prefix Sum + DFS

**Logic:**
Step 1: Use a hash map to store the frequency of prefix sums encountered along the current path.

Step 2: Initialize the map with `{0: 1}` to account for paths starting from the root.

Step 3: During DFS, maintain the `current_sum` from the root to the current node.

Step 4: Check if `current_sum - targetSum` exists in the map; if so, add its frequency to the total count.

Step 5: Update the map with the `current_sum`, recurse into children, and then decrement the frequency (backtrack).

```python
from collections import defaultdict
class Solution:
    def pathSum(self, root: TreeNode, targetSum: int) -> int:
        self.count = 0
        prefix_sums = defaultdict(int)
        prefix_sums[0] = 1
        def dfs(node, curr_sum):
            if not node: return
            curr_sum += node.val
            self.count += prefix_sums[curr_sum - targetSum]
            prefix_sums[curr_sum] += 1
            dfs(node.left, curr_sum)
            dfs(node.right, curr_sum)
            prefix_sums[curr_sum] -= 1
        dfs(root, 0)
        return self.count
```

### [Arrays] Problem 42: Trapping Rain Water

**Algorithm & Technique:** Two Pointers

**Logic:**
Step 1: Initialize two pointers, `left` at the start and `right` at the end of the array.

Step 2: Maintain `left_max` and `right_max` variables.

Step 3: Move the pointer pointing to the smaller height.

Step 4: Update the corresponding max height variable.

Step 5: The water trapped at the current position is the difference between the max height and current height.

Step 6: Accumulate the trapped water until pointers meet.

```python
class Solution:
    def trap(self, height: list[int]) -> int:
        if not height: return 0
        l, r = 0, len(height) - 1
        left_max, right_max = height[l], height[r]
        res = 0
        while l < r:
            if left_max < right_max:
                l += 1
                left_max = max(left_max, height[l])
                res += left_max - height[l]
            else:
                r -= 1
                right_max = max(right_max, height[r])
                res += right_max - height[r]
        return res
```
